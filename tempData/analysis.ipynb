```python
def get_simMat(row, decisionMat, counterMat):
    #only fill in the upper triangle
    leftEmojiIndex = list(plotInfo.loc[plotInfo.emoji_metadata_row == row.leftEmoji, "plot_index"])[0];
    middleEmojiIndex = list(plotInfo.loc[plotInfo.emoji_metadata_row == row.middleEmoji, "plot_index"])[0];
    rightEmojiIndex = list(plotInfo.loc[plotInfo.emoji_metadata_row == row.rightEmoji, "plot_index"])[0];

    if leftEmojiIndex < middleEmojiIndex:
        smallerIndexEmoji = row.leftEmoji
        largerIndexEmoji = row.middleEmoji
    elif leftEmojiIndex > middleEmojiIndex:
        smallerIndexEmoji = row.middleEmoji
        largerIndexEmoji = row.leftEmoji
    counterMat.loc[smallerIndexEmoji, largerIndexEmoji] += 1
    if row.choicePos == "rightEmoji":
        decisionMat.loc[smallerIndexEmoji, largerIndexEmoji] += 1

    if leftEmojiIndex < rightEmojiIndex:
        smallerIndexEmoji = row.leftEmoji
        largerIndexEmoji = row.rightEmoji
    elif leftEmojiIndex > rightEmojiIndex:
        smallerIndexEmoji = row.rightEmoji
        largerIndexEmoji = row.leftEmoji
        
    counterMat.loc[smallerIndexEmoji, largerIndexEmoji] += 1
    if row.choicePos == "middleEmoji":
        decisionMat.loc[smallerIndexEmoji, largerIndexEmoji] += 1

    if middleEmojiIndex < rightEmojiIndex:
        smallerIndexEmoji = row.middleEmoji
        largerIndexEmoji = row.rightEmoji
    elif middleEmojiIndex > rightEmojiIndex:
        smallerIndexEmoji = row.rightEmoji
        largerIndexEmoji = row.middleEmoji
    counterMat.loc[smallerIndexEmoji, largerIndexEmoji] += 1
    if row.choicePos == "leftEmoji":
        decisionMat.loc[smallerIndexEmoji, largerIndexEmoji] += 1

    return    
```
